## Opening an Image

```python
from PIL import Image
im = Image.open("img.jpeg")
im
```

## Features

```python
im.format , im.size , im.mode
```

## Show Image

```python
print(im.show())
```

## Crop Image

```python
box = (100, 100, 400, 400)
region = im.crop(box)
region
```

## Resize

```python
out = im.resize((128, 128))
out
```

## Rotate

```python
out = im.rotate(45)
out
```

## Transpose

```python
# FLIP_LEFT_RIGHT

out = im.transpose(Image.FLIP_LEFT_RIGHT)
out
```

```python
# FLIP_TOP_BOTTOM

out = im.transpose(Image.FLIP_TOP_BOTTOM)
out
```

```python
# ROTATE_90

out = im.transpose(Image.ROTATE_90)
out
```

```python
# ROTATE_180

out = im.transpose(Image.ROTATE_180)
out
```

## GreyScale

```python
out = im.convert("L")
out
```

## ImageFilter

```python
from PIL import ImageFilter
```

```python
# BLUR

blurred_image = im.filter(ImageFilter.BLUR)
blurred_image
```

```python
# GaussianBlur

blurred_image = im.filter(ImageFilter.GaussianBlur)
blurred_image
```

```python
# BoxBlur

blurred_image = im.filter(ImageFilter.BoxBlur(9))
blurred_image
```

## ImageEnhance

```python
from PIL import ImageEnhance

enh = ImageEnhance.Contrast(im)
enh.enhance(1.3)
```

## Converting type of image

```python

converted = Image.open("img.png")
converted.save('converted.jpeg', 'jpeg')  
```

## ImageDraw

```python
from PIL import Image, ImageDraw

image = Image.new('RGBA', (100, 100), 'white')    
drawing = ImageDraw.Draw(image)

drawing.rectangle((5, 5, 50, 50), fill='red')
 
image.save('drawing.png')
```
